```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('VN_House_price_Clean2.csv')
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df = df.dropna(subset=["Price", "City"])
                                         
top_cities = df["City"].value_counts().head(10).index
df_plot = df[df["City"].isin(top_cities)]

plt.figure(figsize=(10,6))
sns.boxplot(data=df_plot, x="City", y="Price", showfliers=False)
plt.title("BOXPLOT GI NH THEO THNH PH (Top 10)", fontweight="bold")
plt.ylabel("Gi (t VN)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# BC QUAN TRNG: c file d liu ca bn
# Thay 'du_lieu_nha_dat.csv' bng ng dn file thc t trn my bn
df = pd.read_csv('VN_House_price_Clean2.csv') 

# To figure vi 2 biu  con (subplots)
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# 1. V Boxplot cho Php l (Legal status)
sns.boxplot(ax=axes[0], x='Legal status', y='Price', data=df, palette='Set2')
axes[0].set_title('Phn b Gi theo Tnh trng Php l', fontsize=14)
axes[0].tick_params(axis='x', rotation=30)
axes[0].set_xlabel('Tnh trng php l')
axes[0].set_ylabel('Gi nh (T VN)')

# 2. V Boxplot cho Ni tht (Furniture state)
sns.boxplot(ax=axes[1], x='Furniture state', y='Price', data=df, palette='Set3')
axes[1].set_title('Phn b Gi theo Tnh trng Ni tht', fontsize=14)
axes[1].tick_params(axis='x', rotation=30)
axes[1].set_xlabel('Mc  ni tht')
axes[1].set_ylabel('Gi nh (T VN)')

plt.tight_layout()
plt.show()
```
